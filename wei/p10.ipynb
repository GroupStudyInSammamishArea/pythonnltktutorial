## 1. WordNet

(1) Not exactly a corpora but really a giant Lexicon.

(2) We can use it to look up words and their meaning according to their parts of speech.

(3) We can also find synonyms, antonyms, and even examples of the word in use.

(4) Application: create/catch plagiarism

## 2. Example

```python
from nltk.corpus import wordnet

syns = wordnet.synsets("program")

print(syns)

print(syns[0])

# Just the word
print(syns[0].lemmas()[0].name())

# synset
print(syns[0].name())

# Definition
print(syns[0].definition())

# Examples
print(syns[0].examples())

synonyms = []
antonyms = []

for syn in wordnet.synsets("good"):
    for l in syn.lemmas():
        #print("l:", l)
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())
            
print(set(synonyms))
print(set(antonyms))

antonyms2 = []

for syn in wordnet.synsets("good"):
    for l in syn.lemmas():
        for a in l.antonyms():
            antonyms2.append(a.name())
            
print(set(antonyms2))

w1 = wordnet.synset("ship.n.01")
w2 = wordnet.synset("boat.n.01")

print(w1.wup_similarity(w2))

w1 = wordnet.synset("ship.n.01")
w2 = wordnet.synset("car.n.01")

print(w1.wup_similarity(w2))

w1 = wordnet.synset("ship.n.01")
w2 = wordnet.synset("cactus.n.01")

print(w1.wup_similarity(w2))
```
